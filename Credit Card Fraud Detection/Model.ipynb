```python
X = df.drop(["Class"],axis=1)
y = df["Class"]
```

```python
from sklearn.model_selection import train_test_split

x_train,x_test,y_train,y_test = train_test_split(X,y,test_size=0.2)
```

```python
from sklearn.svm import SVC

svm = SVC()
```

```python
svm.fit(x_train,y_train)
```

```python
y_prd = svm.predict(x_test)
```

```python
from sklearn.metrics import accuracy_score

accuracy_score(y_test,y_prd)
```

```python
XC = np.concatenate((x_train,x_test))
YC = np.concatenate((y_train,y_test))
```

```python
from sklearn.model_selection import cross_val_score

cv = cross_val_score(svm,XC,YC,cv=20)
```

```python
cv.mean()
```

```python
from sklearn.linear_model import LogisticRegression

lr = LogisticRegression()
```

```python
lr.fit(x_train,y_train)
```

```python
y_prd2 = lr.predict(x_test)
```

```python
accuracy_score(y_test,y_prd2)
```
